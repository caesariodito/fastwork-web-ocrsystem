```python
import pandas as pd
```

```python
df = pd.read_csv('./data/data_gambar_pln.csv', header=None, names=['link'])
df
```

```python
import urllib.request

errors = []
for i in range(81,len(df)):
    try:
        urllib.request.urlretrieve(df['link'][i], f'./data/images/gambar_pln_{i}.jpg')
        print(f'gambar_pln_{i}.jpg downloaded')
    except Exception as e:
        print(f'gambar_pln_{i}.jpg failed to download')
        print(e)
        errors.append(df['link'][i])
```
